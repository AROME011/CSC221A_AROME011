```python
class Money:
  def __init__(self, value):
    self.value = value

  def __str__(self):
    return f"${self.value:.2f}"


class Bills(Money):
  def __init__(self, value):
    super().__init__(value)

class One(Bills):
  def __init__(self):
    super().__init__(1.0)

class Five(Bills):
  def __init__(self):
    super().__init__(5.0)

class Ten(Bills):
  def __init__(self):
    super().__init__(10.0)

class Twenty(Bills):
  def __init__(self):
    super().__init__(20.0)

class Fifty(Bills):
  def __init__(self):
    super().__init__(50.0)


class Coins(Money):
  def __init__(self, value):
    super().__init__(value)

class Penny(Coins):
  def __init__(self):
    super().__init__(0.01)

class Nickel(Coins):
  def __init__(self):
    super().__init__(0.05)

class Dime(Coins):
  def __init__(self):
    super().__init__(0.10)

class Quarter(Coins):
  def __init__(self):
    super().__init__(0.25)

wallet1 = [Fifty(), Ten(), One(), Dime(), Penny()]
wallet2 = [Twenty(), Twenty(), Five(), Five(), Quarter(), Nickel()]
wallet3 = [Fifty(), Twenty(), Ten(), Five(), One(), Dime()]

def counting_total(wallet):
  total = 0
  for money in wallet:
    total += money.value
  return total

def show_wallet(wallet):
  for money in wallet:
    print(money)

print("Wallet 1 contains:")
show_wallet(wallet1)
print(f"The total value is:\n${counting_total(wallet1):.2f}")

print("\nWallet 2 contains:")
show_wallet(wallet2)
print(f"The total value is:\n${counting_total(wallet2):.2f}")

print("\nWallet 3 contains:")
show_wallet(wallet3)
print(f"The total value is:\n${counting_total(wallet3):.2f}")


```

```python
class Teams:
    def __init__(self, name):
        self.name = name
        self.wins = 0
        self.losses = 0

    def record_win(self):
        self.wins = self.wins + 1

    def record_loss(self):
        self.losses = self.losses + 1

    def get_record(self):
        return f"{self.name} Record: {self.wins} - {self.losses}"


class Gameplay:
    def __init__(self, home_team, away_team, home_team_won):
        self.home_team = home_team
        self.away_team = away_team
        self.home_team_won = home_team_won

    def play_game(self):
        if self.home_team_won:
            self.home_team.record_win()
            self.away_team.record_loss()
        else:
            self.home_team.record_loss()
            self.away_team.record_win()

    def results_game(self):
        if self.home_team_won:
          winner = self.home_team.name
        else:
          winner = self.away_team.name
        if self.home_team_won:
          loser = self.away_team.name
        else:
          loser = self.home_team.name
        result = f"{winner} defeated {loser}"
        return result


team1 = Teams("Plattsburgh Cardinals")
team2 = Teams("Akron Tigers")

game1 = Gameplay(team1, team2, home_team_won=True)
game2 = Gameplay(team1, team2, home_team_won=False)
game3 = Gameplay(team1, team2, home_team_won=False)
game4 = Gameplay(team1, team2, home_team_won=True)
game5 = Gameplay(team1, team2, home_team_won=False)
game6 = Gameplay(team1, team2, home_team_won=True)
game7 = Gameplay(team1, team2, home_team_won=True)

game1.play_game()
game2.play_game()
game3.play_game()
game4.play_game()
game5.play_game()
game6.play_game()
game7.play_game()

print(f"Game 1: {game1.results_game()}")
print(f"\nGame 2: {game2.results_game()}")
print(f"\nGame 3: {game3.results_game()}")
print(f"\nGame 4: {game4.results_game()}")
print(f"\nGame 5: {game5.results_game()}")
print(f"\nGame 6: {game6.results_game()}")
print(f"\nGame 7: {game7.results_game()}")


print(f"\n{team1.get_record()}")
print(team2.get_record())
```
